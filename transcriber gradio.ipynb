```python
import os
import whisper
import gradio as gr
import subprocess

def download_audio_with_ytdlp(video_url):
    """Download audio from a YouTube video using yt-dlp."""
    try:
        output_file = "audio.mp3"
        command = [
            "yt-dlp",
            "--extract-audio",
            "--audio-format", "mp3",
            "--output", output_file,
            video_url
        ]
        subprocess.run(command, check=True)
        return output_file
    except subprocess.CalledProcessError as e:
        return f"Error: {e}"

def transcribe_audio(file_path, language="en"):
    """Transcribe audio using OpenAI's Whisper."""
    try:
        model = whisper.load_model("base")
        result = model.transcribe(file_path, language=language)
        return result["text"]
    except Exception as e:
        return f"Error during transcription: {e}"

def process_youtube_video(video_url, language):
    """Process YouTube video: Download audio and transcribe."""
    audio_file = download_audio_with_ytdlp(video_url)
    if "Error" in audio_file:
        return audio_file
    transcription = transcribe_audio(audio_file, language)
    os.remove(audio_file)  # Clean up the audio file
    return transcription

# Gradio Interface
def gradio_interface(video_url, language):
    return process_youtube_video(video_url, language)

interface = gr.Interface(
    fn=gradio_interface,
    inputs=[
        gr.Textbox(label="YouTube Video URL"),
        gr.Radio(["en", "es", "fr", "de"], value="en", label="Language")
    ],
    outputs="text",
    title="YouTube Audio Transcriber",
    description="Enter a YouTube video URL and select the language to transcribe its audio using Whisper."
)

if __name__ == "__main__":
    interface.launch()
```
